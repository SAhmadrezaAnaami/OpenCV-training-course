### **Loading Images**

```python
# Load an image using 'imread' specifying the path to image
image = cv2.imread('./images/flowers.jpg')
```

### **Displaying Images**

```python
from matplotlib import pyplot as plt

#Show the image with matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Let's create a simple function to make displaying our images simpler and easier

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Let's test it out
imshow("Displaying Our First Image", image)
```

### **Saving Images**

```python
# Simply use 'imwrite' specificing the file name and the image to be saved
cv2.imwrite('output.jpg', image)
```

```python
# Or save it as a PNG (Portable Network Graphics) which is a Lossless bitmap image format
cv2.imwrite('output.png', image)
```

## calculate the Resolution of lena image in inch
#### hint:

They got tired of their stock of usual test images, dull stuff dating back to television standards work in the early 1960s. They wanted something glossy to ensure good output dynamic range, and they wanted a human face. Just then, somebody happened to walk in with a recent issue of Playboy. The engineers tore away the top third of the centerfold so they could wrap it around the drum of their Muirhead wirephoto scanner, which they had outfitted with analog-to-digital converters (one each for the red, green, and blue channels) and a Hewlett Packard 2100 minicomputer. The Muirhead had a fixed resolution of ***100 lines per inch*** and the engineers wanted a 512×512 image, so they limited the scan to the top 5.12 inches of the picture, effectively cropping it at the subject's shoulders.

```python
def calculate_resolution(image_path, standard_dpi=100): # DPI (dots per inch) 
    # Open the image
    image = image = cv2.imread(image_path ,0)
    
    # Get the width and height of the image in pixels
    width_pixels, height_pixels = image.shape
    
    # Calculate the resolution in inches using a standard DPI value
    resolution_inches = (width_pixels / standard_dpi, height_pixels / standard_dpi)
    return resolution_inches

# Example usage
image_path = "./images/lena512.png"  # Replace this with the path to your image
resolution_inches = calculate_resolution(image_path)
print("Resolution (in inches):", resolution_inches)
```
